# Different types of Activation Function

Sigmoid

Hyperbolic Tangent (tanh)

Rectified Linear Units (ReLU)

Leaky ReLU

Softmax

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sigmoid Activation Function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=5000
print(sigmoid(x))
```

```python
x = np.linspace(-500, 500, 11)
print(x)
```

```python
p = sigmoid(x)
print(p)
```

```python
plt.xlabel('x')
plt.ylabel('sigmoid')
plt.plot(x,p)
plt.show()
```

```python
#TanH activation FUnction
def tanh(x):
    return np.tanh(x)
```

```python
x = 600
```

```python
print(tanh(x))
```

```python
x =np.linspace(-500, 500, 11)
print(x)
```

```python
p = tanh(x)
print(p)
```

```python
plt.xlabel('x')
plt.ylabel('tanh')
plt.plot(x,p)
plt.show()
```

```python
 #Relu 
def relu(x):
    return np.maximum(0,x)
```

```python
x= 600
relu(x)
```

```python
x= np.linspace(-500, 500, 11)
print(x)
```

```python
p = relu(x)
print(p)
```

```python
plt.xlabel('x')
plt.ylabel('relu')
plt.plot(x,p)
plt.show()
```

```python
#leaky relu
def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha*x, x)
```

```python
x =800
leaky_relu(x)
```

```python
x = np.linspace(-500, 500,11)
print(x)
```

```python
p = leaky_relu(x)
print(p)
```

```python
plt.xlabel('x')
plt.ylabel('leaky_relu')
plt.plot(x,p)
plt.show()
```

```python
#softmax
def softmax(x):
    exp_x = np.exp(x-np.max(x))
    return exp_x / np.sum(exp_x, axis=0)
```

```python
x = 600
print(softmax(x))
```

```python
x = np.linspace(10000,-10000,11)
print(x)
```

```python
p =softmax(x)
print(p)
```

```python
x = np.array([2.0, 1.0,0.5])
print(x)
```

```python
p = softmax(x)
p
```

```python
np.sum(p)
```

```python
plt.xlabel('x')
plt.ylabel('softmax')
plt.plot(x,p)
plt.show()
```
