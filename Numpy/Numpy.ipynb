```python
x_sin=np.arange(0,2*np.pi, 0.1)
y_sin=np.sin(x_sin)
print(y_sin)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_sin,y_sin)
plt.title('sin curve')
plt.show()
```

```python
x_cos=np.arange(0,2*np.pi,0.1)
y_cos=np.cos(x_cos)

plt.figure(figsize=(6,6))
plt.plot(x_cos,y_cos)
plt.show()
```

```python
x_tan=np.arange(0,2*np.pi,0.1)
y_tan=np.tan(x_tan)

plt.figure(figsize=(6,6))
plt.plot(x_tan,y_tan)
plt.title('TAN CURVE')
plt.show()
```

```python
x_cot=np.arange(0,2*np.pi,0.1)
y_cot=1/np.tan(x_cot)

plt.figure(figsize=(6,6))
plt.plot(x_cot,y_cot,'b-')
plt.title("COT CURVE")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.subplot(2,2,1)
plt.plot(x_sin,y_sin, 'r-')
plt.title('SIN CURVE')

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos, 'g-')
plt.title('cos CURVE')

plt.subplot(2,2,3)
plt.plot(x_tan,y_tan, 'b-')
plt.title('tan CURVE')

plt.subplot(2,2,4)
plt.plot(x_cot,y_cot, 'r-')
plt.title('cot CURVE')

plt.show()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
sns.displot(np.random.binomial(n=10, p=0.5, size=8), kind='kde')
plt.title("binomial ")
plt.xlabel('number of success head')
plt.ylabel('frequency')
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate call counts using Poisson distribution
calls_per_hour = random.poisson(lam=5, size=1000)

# Plotting the result
sns.histplot(calls_per_hour, bins=range(0, 15), kde=False, color='orange')
plt.title("Poisson Distribution: Calls Received Per Hour (=5)")
plt.xlabel("Number of Calls in 1 Hour")
plt.ylabel("Frequency (Out of 1000 Hours)")
plt.show()
```
