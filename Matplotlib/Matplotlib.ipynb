```python
import numpy as np
import matplotlib.pyplot as plt

Temp_BNG = np.random.randint(20, 40, size=10)
Hum_BNG = np.random.randint(10, 50, size=10)

Temp_DVG = np.random.randint(20, 40, size=10)
Hum_DVG = np.random.randint(10, 50, size=10)

print(f"for Bangalore: {Temp_BNG}, {Hum_BNG}")
print(f"for Davanagere: {Temp_DVG}, {Hum_DVG}")

plt.figure(figsize=(12,6))
plt.scatter(Temp_BNG, Hum_BNG, color='blue', s=200, label='Bangalore')
plt.scatter(Temp_DVG, Hum_DVG, color='green', marker='v', s=100, label='Davanagere')

plt.xticks(np.arange(0, 60, 5))
plt.yticks(np.arange(0, 60, 5))

plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.title('Weather Comparison: Bangalore vs Davanagere')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

subject = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks1 = [89, 90, 45, 78, 90]
marks2 = [85, 88, 50, 70, 95]

subject_len = np.arange(len(subject))   # → [0, 1, 2, 3, 4]
width = 0.4

plt.bar(subject_len, marks1, width=width, label='Student A')                  # Bars at [0,1,2,3,4]
plt.bar(subject_len + width, marks2, width=width, label='Student B')          # Bars at [0.4,1.4,...]

plt.xticks(subject_len + width / 2, subject)  # Shift x-axis labels to be centered between grouped bars

plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Comparison of Marks Between Two Students')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
roll = [1, 2, 3, 4, 5]
number = [11, 12, 13, 14, 15]
marks = [80, 65, 70, 75, 90]
subject = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']

plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
plt.plot(roll, marks, marker='o')
plt.title('Roll vs Marks')
plt.xlabel('Roll')
plt.ylabel('Marks')

plt.subplot(2, 2, 2)
plt.plot(number, marks, color='orange', marker='s')
plt.title('Number vs Marks (Orange)')

plt.subplot(2, 2, 3)
plt.plot(number, marks, marker='^', linestyle='--')
plt.title('Number vs Marks (Default)')

plt.subplot(2, 2, 4)
plt.bar(subject, marks, edgecolor='white', color='skyblue')
plt.title('Subject-wise Marks')
plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```
