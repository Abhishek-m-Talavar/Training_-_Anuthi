```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
roll=np.arange(1,16)
marks=np.random.randint(30,60, size=15)
df1=pd.DataFrame({'Roll':roll,'Marks':marks})
print(df1)
```

```python
sns.lineplot(x='Roll',y='Marks',data=df1)
plt.title('Student Marks')
plt.show()
```

```python
flight_data = sns.load_dataset('flights')
flight_data.head(3)
```

```python
flight_data = flight_data.pivot(index='month', columns='year', values='passengers')

flight_data.head()
```

```python
plt.figure(figsize=(12,8))
x=sns.heatmap(flight_data)
x
```

```python
plt.figure(figsize=(14,8))
x=sns.heatmap(flight_data, annot=True ,fmt='d',linecolor='k',linewidths='4',
                 cmap='Blues')
x
```
